```python
#importing libraries
import os
import openai
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.schema import Document
from langchain.document_loaders import TextLoader
from pinecone import ServerlessSpec

```

```python
#importing the data
book = '/content/drive/MyDrive/10academy/rags/alice_wonderland_book.txt'
with open(book, 'r') as file:
    book_content = file.read()

print(book_content)
```

```python
loader = TextLoader(book_content)
data = loader.load()
```

```python
class Document:
    def __init__(self, page_content, metadata=None):
        self.page_content = str(page_content)  # Ensure page_content is a string
        self.metadata = metadata or {}

#splitting the data using text splitter
data = Document(book_content, metadata={"some_key": "some_value"})

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=500, add_start_index=True)
data1 = text_splitter.split_documents([data])
```

```python
print(data1)
```

```python
from pinecone import Pinecone

# initialize connection to pinecone
api_key = os.environ.get('pinecone_key')

# configure client
pc = Pinecone(api_key=api_key)
```

```python
import time

index_name = 'rag-chatbot'
existing_indexes = [
    index_info["name"] for index_info in pc.list_indexes()
]

# creating index
if index_name not in existing_indexes:
    # if does not exist, create index
    pc.create_index(
        index_name,
        dimension=1536,  # dimensionality of ada 002
        metric='cosine',
        spec=ServerlessSpec(
        cloud="aws",
        region="us-west-2"
    )

    )
    # wait for index to be initialized
    while not pc.describe_index(index_name).status['ready']:
        time.sleep(1)

# connect to index
index = pc.Index(index_name)
time.sleep(1)
# view index stats
index.describe_index_stats()
```

```python
#initializing the api key
key3 = os.environ.get('openai_apikey')
openai.api_key=key3
```

```python
#embedding doc
embed_model = OpenAIEmbeddings(model="text-embedding-ada-002", api_key = openai.api_key)
```

```python
# Initialize lists to store text, metadata, and embeddings
texts = []
metadatas = []
embeddings = []
```

```python
print(data)
```

```python
#splitting the doc to chunks
chunk_size = 500
chunks = [data1[i:i+chunk_size] for i in range(0, len(data1), chunk_size)]
print(chunks[0])
```

```python
from tqdm.auto import tqdm
import pandas as pd
from uuid import uuid4

# Iterate over the chunks and create embeddings
for i, chunk in tqdm(enumerate(chunks)):
    # Generate a unique ID for each chunk
    chunk_id = str(uuid4())

    # Embed the chunk using OpenAI model
    chunk_embedding = embed_model.embed_documents(book_content)  # Use your OpenAI embedding model here

    # Append data to lists
    texts.append(chunk)
    metadatas.append({'chunk_id': chunk_id, 'chunk_number': i})
    embeddings.append(chunk_embedding)

    # If we reach the batch limit, upsert to Pinecone
    if len(texts) >= chunk_size:
        ids = [str(uuid4()) for _ in range(len(texts))]
        index.upsert(vectors=zip(ids, embeddings, metadatas))
        texts = []
        metadatas = []
        embeddings = []

# Handle any remaining chunks
if len(texts) > 0:
    ids = [str(uuid4()) for _ in range(len(texts))]
    index.upsert(vectors=zip(ids, embeddings, metadatas))

```
